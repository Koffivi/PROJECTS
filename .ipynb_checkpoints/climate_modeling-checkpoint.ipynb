## Task 1: Energy Balance Models (EBM version 1)
This ```EBM01``` is for to the global mean temperature of the
earth by solving the following equation:
$$S(1-\alpha)=4\varepsilon \sigma T^{4}$$

where:
1.  T = Global mean temperature of the earth (dependent variable)
2.  S = Solar radiation (variable; default value = 1372 W m -2 )
3.  α = Planetary albedo (variable; default value = 0.3)
4.  ε = Emissivity of the atmosphere (variable; default value =0.6)
5.  σ = Stephan-Boltzmann constant (constant value = 5.67 x 10 -8 )


````python
def model1(S=1372,alpha=0.3,eps=0.6,sigma=5.67*10**(-8)):
    return (S*(1-alpha)/(4*eps*sigma))**(1/4)
````

### Plots of the curve of the Earth's temperature when we vary  solar radiation, albedo and emissivity

````python

# plot variation of Solar radiation
plt.figure(figsize=(10, 20))
plt.subplot(3,1,1)
plt.plot(Solar_radiation,model1(S=Solar_radiation),color="red")
plt.xlabel('Solar radiation', fontsize = 22)
plt.ylabel('Temperature (K)', fontsize = 22)
plt.title('Change of solar radiation between {} and {}'.format(Solar_radiation[0],Solar_radiation[-1]),fontsize = 15)
plt.grid(True)
plt.show()


# plot variation of Planetary albedo
plt.figure(figsize=(10, 20))
plt.subplot(3,1,2)
plt.plot(Planetary_albedo,model1(alpha=Planetary_albedo),color="red")
plt.xlabel('Planetary albedo',fontsize = 22)
plt.ylabel('Temperature (K)',fontsize = 22)
plt.title('Change of Planetary albedo between {} and {}'.format(Planetary_albedo[0],Planetary_albedo[-1]),fontsize = 15)
plt.grid(True)
plt.show()


# plot variation of Emissivity
plt.figure(figsize=(10, 20))
plt.subplot(3,1,3)
plt.plot(Emissivity,model1(eps=Emissivity),color="red")
plt.xlabel('Emissivity', fontsize = 22)
plt.ylabel('Temperature (K)',fontsize = 22)
plt.title('Change of Emissivity between {} and {}'.format(Emissivity[0],Emissivity[-1]) ,fontsize = 15)
plt.grid(True)
plt.show()
````

## Task 2: Energy Balance Models (EBM version 2)

$$ C_p \dfrac{\partial T}{\partial t} = S(1-\alpha) - 4\epsilon \sigma T^4$$
$$\dfrac{\partial T}{\partial t} =\dfrac{S(1-\alpha) - 4\epsilon \sigma T^4}{C_p}$$


````python
def dTdt(t , T):   
    Cp = 1.05*10**21
    S=1372
    alpha=0.3
    eps=0.6
    sigma=5.67*10**(-8)
    return  (S*(1-alpha)-4*eps*sigma*T**4)/Cp
````

````python
from scipy.integrate import odeint
````

````python
T0=model1()
t=np.linspace(0,200,1000)
````

````python
sol = odeint(dTdt, y0=T0, t=t, tfirst=True)
````

````python
# plt.plot(t,sol)
````

````python
dt=1
S=1372
a=0.3
e=0.6
Cp=100
o=5.67*(10**(-8))
T=np.zeros(200)
for i in range(199):
   
    Fn=(S*(1-a)-4*e*o*(T[i])**4)/Cp
    T[i+1]=T[i]+Fn*dt
# T
t=np.linspace(0,200,1000)
````

````python
plt.figure(figsize=(10,8))
plt.plot(T,label="EBM2")
plt.plot(t,sol,label="EBM1")

plt.xlabel('Time', fontsize = 22)
plt.ylabel('Temperature (K)',fontsize = 22)
plt.title('Comparaison between EBM1 and EBM2'.format(Emissivity[0],Emissivity[-1]) ,fontsize = 15)
plt.legend()
plt.grid(True)
plt.show()
````

We define a function named Temperatureinitialvalues to get the records of the  initial temperatures that we want to use to see the variations in the plot.

Note: the function is dynamic

````python
def Temperatureinitialvalues():
    nb=eval(input("Enter the number of initial temperature for the test(eg:5 ): "))
    lalist=[]
    s=0
    for i in range(nb):
        s=eval(input("Enter the number {} value of initial temperature in Kelvin (eg:273) : ".format(i+1)))
        lalist.append(s)
    return lalist
    
````

````python
temperature_initial_values_list=Temperatureinitialvalues()
````

````python
plt.figure(figsize=(10,8))
for j in temperature_initial_values_list:
    dt=1
    S=1372
    a=0.3
    e=0.6
    Cp=100
    o=5.67*(10**(-8))
    T=[j*i/i for i in range(1,201)]
    for i in range(199):
        Fn=(S*(1-a)-4*e*o*(T[i])**4)/Cp
        T[i+1]=T[i]+Fn*dt
    # T
    t=np.linspace(0,200,1000)
    plt.plot(T,label= "Initial Temperature ={} ".format(j))
    plt.legend()
    plt.grid(True)
plt.xlabel('Time', fontsize = 22)
plt.ylabel('Temperature (K)',fontsize = 22)
plt.title("Variation of Temperature in time with different values of Initial temperature without albedo feedback",fontsize = 15)
````

We define a function named Albedo_list_generator to get the records of the albedo  that we want to use to see the variations in the plot.

Note: the function is dynamic

````python
def Albedo_list_generator():
    nb=eval(input("Enter the number of values of albedo (eg:2,3,4... ): "))
    lalist=[]
    s=0
    for i in range(nb):
        s=eval(input("Enter the number {} value the albedo (eg:the real value is 0.3) : ".format(i+1)))
        lalist.append(s)
    return lalist
````

Exemple:

Enter the number of values of albedo (eg:2,3,4... ): 2

Enter the number 1 value the albedo (eg:the real value is 0.3) : 0.3

Enter the number 2 value the albedo (eg:the real value is 0.3) : 0.5


````python
Albedo_list=Albedo_list_generator()
````

````python
plt.figure(figsize=(18,13))
for a in Albedo_list:
    for j in temperature_initial_values_list:
        dt=1
        S=1372
        e=0.6
        Cp=100
        o=5.67*(10**(-8))
        T=[j*i/i for i in range(1,201)]
        for i in range(199):
            Fn=(S*(1-a)-4*e*o*(T[i])**4)/Cp
            T[i+1]=T[i]+Fn*dt
        # T
        t=np.linspace(0,200,1000)
        plt.plot(T,label= "Initial Temperature ={} and albedo={} ".format(j,a))
        plt.legend()
        plt.grid(True)
plt.xlabel('Time', fontsize = 22)
plt.ylabel('Temperature (K)',fontsize = 22)
plt.title("Variation of Temperature in time with different values of Initial temperature and albedo",fontsize = 15)

    
````
